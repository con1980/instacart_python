## Task 5
### Determine a suitable way to combine the orders_products_combined dataframe with your products data set. 

#### Most suitable way to combine the two dataframes is with the merge() function on the product_id column which they both have in common.

```python
#Merge the two dataframes
df_ords_prods_merge = df_ords_prods_comb.merge(df_prods, on = 'product_id', indicator = True, how = 'inner')
```

```python
#check if merge completed accordingly
df_ords_prods_merge.head()
```
