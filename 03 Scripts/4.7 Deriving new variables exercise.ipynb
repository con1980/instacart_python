## 2.
### Derive variable in dataframe with if statement

```python
#write first million rows in new dataframe df
df=df_ords_prods_merge[:1000000]
```

```python
#Define function to evaluate price ranges for products
def price_label(row):
    if row['prices']<=5:
        return 'Low-range Product'
    elif (row['prices']>5) and (row['prices']<=15):
        return 'Mid-range Product'
    elif row['prices']>15:
        return 'High-range Product'
    else: return 'not enough data'
    
```

```python
#Apply function to dataframe
df['price_range']=df.apply(price_label,axis=1)
```

```python
# check dataframe if execution went as wished
df.head()
```

```python
#Overview of which price ranges appear how many times in dataframe
df['price_range'].value_counts(dropna=False)
```

## 3.
### ### Derive variable in dataframe with loc() function

```python
#loc function for high-range products
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
#loc function for Mid-range products
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product' 
```

```python
#loc function for Low-range products
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
#Overview of which price ranges appear how many times in dataframe
df['price_range_loc'].value_counts(dropna=False)
```

## 5.
### Derive variable in dataframe df_ords_prods_merge with for loop

```python
#check which day is the busiest and least busy.
df_ords_prods_merge['orders_day_of_week'].value_counts(dropna=False)
```

#### from our project brief we know that 0 is Saturday and 4 is Wednesday.

```python
# make new list called result
result = []

#Loop through whole dataframe in column 'orders_day_of_week' 
#and write the corresponding text for when if clause is true in the list
for value in df_ords_prods_merge["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
# create new column with name 'busiest day' and write values from 'result' list in that column
df_ords_prods_merge['busiest_day']=result
```

```python
# see overview of count of the three groups, 'Regurlarly busy', 'Busiest day', and 'Least busy'
df_ords_prods_merge['busiest_day'].value_counts(dropna=False)
```

```python
#check if shape is the same
df_ords_prods_merge['busiest_day'].shape==df_ords_prods_merge['orders_day_of_week'].shape
```

# 6. Task Step 2
### Suppose your clients have changed their minds about the labels you created in your “busiest_day” column. Now, they want “Busiest day” to become “Busiest days” (plural). This label should correspond with the two busiest days of the week as opposed to the single busiest day. At the same time, they’d also like to know the two slowest days. Create a new column for this using a suitable method.


```python
#check which two days are the busiest and which the least busy.
df_ords_prods_merge['orders_day_of_week'].value_counts(dropna=False)
```

### It appears that Saturday (0) and Sunday (1) are the two most busy days. And Wednesday (4) and Tuesday (3) the least busy days.

```python
# make new list called 'busiest_days_list'
busiest_days_list = []

#Loop through whole dataframe in column 'orders_day_of_week' 
#and write the corresponding text for when if clause is true in the list
for value in df_ords_prods_merge["orders_day_of_week"]:
  if value == 0 or value == 1:
    busiest_days_list.append("Busiest days")
  elif value == 4 or value == 3:
    busiest_days_list.append("Least busy days")
  else:
    busiest_days_list.append("Regularly busy days")
```

```python
busiest_days_list
```

```python
# create new column with name 'busiest_days' and write values from 'busiest_days_list' list in that column
df_ords_prods_merge['busiest_days']=busiest_days_list
```

```python
#check outcome
df_ords_prods_merge[['orders_day_of_week','busiest_days']].head(30)
```

# 8. Task Step 4
### When too many users make Instacart orders at the same time, the app freezes. The senior technical officer at Instacart wants you to identify the busiest hours of the day. Rather than by hour, they want periods of time labeled “Most orders,” “Average orders,” and “Fewest orders.” Create a new column containing these labels called “busiest_period_of_day.”

```python
#check which hours are the busiest and which the least busy.
df_ords_prods_merge['order_hour_of_day'].value_counts(dropna=False)
```

```python
# make new list called 'busiest_period'
busiest_period = []

#Loop through whole dataframe in column 'order_hour_of_day' 
#and write the corresponding text for when if clause is true in the list
for value in df_ords_prods_merge["order_hour_of_day"]:
  if value in (10,11,14,15,13,12,16,9):
    busiest_period.append("Most orders")
  elif value in (17,8,18,19,20,7,21,22):
    busiest_period.append("Average orders")
  else:
    busiest_period.append("Fewest orders")
```

```python
#check outcome
busiest_period
```

```python
# create new column with name 'busiest_period_of_day' and write values from 'busiest_period' list in that column
df_ords_prods_merge['busiest_period_of_day']=busiest_period
```

```python
#check outcome
df_ords_prods_merge[['order_hour_of_day','busiest_period_of_day']].head(20)
```
