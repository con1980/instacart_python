```python
#Group by user_id and apply transform function on order_number with max argument to find maximum orders
#write in new column 'max_order'
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform('max')
```

# 5. Task step 6
### The team now wants to target different types of spenders in their marketing campaigns. This can be achieved by looking at the prices of the items people are buying. Create a spending flag for each user based on the average price across all their orders using the following criteria:
### If the mean of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender.”
### If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a “High spender.”

```python
#Group by user_id and apply transform function on prices with mean argument to find mean prices
#write in new column 'mean_prices'
ords_prods_merge['mean_prices'] = ords_prods_merge.groupby(['user_id'])['prices'].transform('mean')
```

```python
#check if execution went accordingly
ords_prods_merge.head(60)
```

```python
#loc function for low spending customers, mean price below 10
ords_prods_merge.loc[ords_prods_merge['mean_prices'] < 10, 'spending_behaviour'] = 'Low spender'
```

```python
#loc function for high spending customers, mean price 10 or above
ords_prods_merge.loc[ords_prods_merge['mean_prices'] >= 10, 'spending_behaviour'] = 'High spender'
```

```python
#check if executed correctly
ords_prods_merge.tail(60)
```

```python
# check if executed correctly
ords_prods_merge.groupby(['user_id','mean_prices']).agg({'spending_behaviour': ['max']}).head(20)
```

# 6. Task step 7
### In order to send relevant notifications to users within the app (for instance, asking users if they want to buy the same item again), the Instacart team wants you to determine frequent versus non-frequent customers. Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column. The criteria for the flag should be as follows:
### If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.”
### If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.”
### If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.”

```python
#Group by user_id and apply transform function on days_since_prior_order with mean argument to find mean days
#write in new column 'mean_prices'
ords_prods_merge['median_days_since_prior_order'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform('mean')
```

```python
ords_prods_merge.head(60)
```

```python
#loc function for non frequent customers, mean days higher 20
ords_prods_merge.loc[ords_prods_merge['median_days_since_prior_order'] > 20, 'order_behaviour'] = 'Non-frequent customer'
```

```python
#loc function for regular customers, mean days between 10 and 20 incl. 20 
ords_prods_merge.loc[(ords_prods_merge['median_days_since_prior_order'] > 10) & (ords_prods_merge['median_days_since_prior_order'] <= 20), 'order_behaviour'] = 'Regular customer'
```

```python
#loc function for frequent customers, mean days lower or equal 10 days
ords_prods_merge.loc[ords_prods_merge['median_days_since_prior_order'] <= 10, 'order_behaviour'] = 'Frequent customer'
```

```python
#Check if all executed correctly by counting the frequency
ords_prods_merge['median_days_since_prior_order'].value_counts(dropna = False)
```
